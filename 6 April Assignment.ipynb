```python
Q3. What is the kernel trick in SVM?
```

ANS-

```python
The kernel trick is a method used in Support Vector Machines (SVM) to transform data that is not linearly separable in a lower dimension
into a higher dimension where it can be linearly separated. The kernel trick is used to map the low-dimensional input space and transform 
it into a higher dimensional space. This allows the SVM algorithm to work well in non-linear data cases using kernel trick.

The SVM kernel could be a function that takes low dimensional input space and transforms it into a better dimensional space, 
i.e., it converts non-separable problems to separable problems. It helps us to deal with non-linear separation problems.
```

```python
Q4. What is the role of support vectors in SVM Explain with example
```

ANS-

```python
Support Vector Machines (SVM) are a set of supervised learning methods used for classification, regression and outliers detection. 
SVM uses the input data points or features called support vectors to maximize the decision boundaries i.e. the space around the hyperplane.
Support vectors are the data points or vectors that are the closest to the hyperplane and which affect the position of the hyperplane.
Since these vectors support the hyperplane, hence called a Support vector. 

For example, consider a dataset with two classes as shown in the image below. The goal is to find a line that separates these two classes.
There are multiple lines that can be drawn to separate these two classes. However, we want to find a line that maximizes the margin between 
these two classes. The margin is defined as the distance between the line and the closest data points from each class. In this case,
we can see that there are three lines that can be drawn to separate these two classes. However, only one line maximizes the margin between
these two classes. This line is called the maximum margin hyperplane (MMH). The data points that lie on this line are called support vectors.
```

```python
Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in
SVM?
```

ANS-

```python
Support Vector Machines (SVM) is a supervised machine learning algorithm that can be used for classification or regression problems.
SVMs are based on the idea of finding a hyperplane that best separates the different classes in the feature space.
The hyperplane is chosen such that it maximizes the margin between the two classes. The margin is defined as the distance between the
hyperplane and the closest data points from each class. There are two types of margins: hard margin and soft margin. Hard margin SVMs 
are used when the data is linearly separable, while soft margin SVMs are used when the data is not linearly separable.

The hyperplane that maximizes the margin is called the maximum margin hyperplane (MMH). The MMH is unique and can be found by solving an
optimization problem.

The marginal plane is a plane that is parallel to the hyperplane and passes through the closest data points from each class.
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y , test_size=0.25 , random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc= SVC(kernel= 'linear')
```

```python
svc.fit(X_train , y_train)
```

```python
svc.coef_
```

```python
y_pred = svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
print(accuracy_score(y_test , y_pred))
print(classification_report(y_test , y_pred))
print(confusion_matrix(y_test , y_pred))
```
